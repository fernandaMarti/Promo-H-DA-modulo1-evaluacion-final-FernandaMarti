```python
class TiendaOnline:
    def __init__(self):
        self.inventario = [] #lista vacia
        self.clientes = {} #diccionario vacío
        self.ventas_totales = 0.0 #iniciado con valor cero y uso del dato float

    def agregar_producto(self, nombre, precio, cantidad):
#. Debe recibir el nombre, precio y cantidad del producto como parámetros
#agrega un producto al inventario o actualiza su cantidad si ya existe                
        for producto in self.inventario:
            if producto['nombre'] == nombre: #itera según el parametro en el inventario
                producto['cantidad'] += cantidad #actualiza la cantidad
                return
        nuevo_producto = {'nombre': nombre, 'precio': precio, 'cantidad': cantidad}
        self.inventario.append(nuevo_producto)
        print(self.inventario)

    def ver_inventario(self):
        #print("Ver Inventario:", self.inventario)
        for producto in self.inventario:
            print(f"Nombre: {producto['nombre']}, Precio: €{producto['precio']}, Cantidad: {producto['cantidad']}")

    def buscar_producto(self, nombre):
        for producto in self.inventario:
            if producto['nombre'] == nombre:
                print("Producto encontrado en el inventario=>")
                print(f"Nombre: {producto['nombre']}, Precio: €{producto['precio']}, Cantidad: {producto['cantidad']}")
                return
        print("El producto es inexistente en el inventario.")

    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario:
            if producto['nombre'] == nombre:
                producto['cantidad'] += cantidad
                return
        print("El producto no existe en el inventario.")

    def eliminar_producto(self, nombre):
        for producto in self.inventario:
            if producto['nombre'] == nombre:
                self.inventario.remove(producto)
                print(f"Producto {nombre} eliminado del inventario.")
                return
        print("Producto no encontrado en el inventario.")

    def calcular_valor_inventario(self):
        valor_total = 0
        for producto in self.inventario:
            valor_total += producto['precio'] * producto['cantidad']
        print(f"El valor total del inventario es: €{valor_total}")

    def realizar_compra(self):
        carrito = []
        while True:
            print("Seleccione un producto del inventario: Camisa, pantalon, falda, vestido")
            self.ver_inventario()
            nombre_producto = input("Escriba el nombre del producto que desea comprar o cancelar para terminar: ")
            if nombre_producto.lower() == 'cancelar':
                break
            for producto in self.inventario:
                if producto['nombre'] == nombre_producto:
                    cantidad = int(input(f"Ingrese la cantidad de {nombre_producto} que desea comprar: "))
                    if cantidad <= producto['cantidad']:
                        carrito.append({'nombre': nombre_producto, 'precio': producto['precio'], 'cantidad': cantidad})
                        producto['cantidad'] -= cantidad
                    else:
                        print("No hay suficiente cantidad del producto disponible.")
                    break
            else:
                print("Producto no encontrado en el inventario.")
        
        total_compra = sum(item['precio'] * item['cantidad'] for item in carrito)
        print(f"El costo total de la compra es: €{total_compra}")
        self.ventas_totales += total_compra


tienda = TiendaOnline()
tienda.agregar_producto('camisa', 20, 40)
tienda.agregar_producto('pantalón', 30, 30)
tienda.agregar_producto('vestido', 50, 15)
tienda.agregar_producto('falda', 35, 10)
tienda.ver_inventario()
tienda.buscar_producto('camisa')
tienda.actualizar_stock('camisa', 100)
tienda.eliminar_producto('vestido')
tienda.calcular_valor_inventario()
tienda.realizar_compra()
print("Ventas totales:", tienda.ventas_totales)
```
